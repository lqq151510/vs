````python
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
np.random.seed(0)
arr=np.random.randn(100)
sns.distplot(arr, bins=10,hist=True, kde=True, rug=True)
````

### 代码解释

以下代码使用 Seaborn 绘制了一个带有核密度估计（KDE）和数据地毯（Rug）的分布图：

```python
np.random.seed(0)
arr = np.random.randn(100)
sns.distplot(arr, bins=10, hist=False, kde=True, rug=True)
```

- `np.random.seed(0)`：设置随机数种子，以确保生成的随机数序列可重复。
- `np.random.randn(100)`：生成 100 个服从标准正态分布（均值为 0，标准差为 1）的随机数，并存储在变量 `arr` 中。
- `sns.distplot(arr, bins=10, hist=False, kde=True, rug=True)`：
  - `arr`：输入的数据数组。
  - `bins=10`：将数据分为 10 个区间（仅在绘制直方图时生效）。
  - `hist=False`：不绘制直方图。
  - `kde=True`：绘制核密度估计曲线。
  - `rug=True`：在图的底部添加数据地毯（Rug），显示每个数据点的位置。
- 返回值是一个图形对象，用于进一步自定义图形。

````python
array=np.random.randint(0,100,500)
sns.distplot(array,hist=False,rug=True)
````

### 代码解释

以下代码使用 Seaborn 绘制了一个双变量散点图（Jointplot）：

```python
dataframe = pd.DataFrame({'x': np.random.randn(500), 'y': np.random.randn(500)})
sns.jointplot(x='x', y='y', data=dataframe, kind='scatter', color='r', size=10, ratio=5, space=1)
```

- `pd.DataFrame({'x': ..., 'y': ...})`：创建一个包含两列（`x` 和 `y`）的 DataFrame，每列包含 500 个随机生成的标准正态分布数据。
- `sns.jointplot(...)`：绘制一个双变量图，展示 `x` 和 `y` 的关系。
  - `x='x', y='y'`：指定横轴和纵轴的数据列。
  - `data=dataframe`：指定数据来源为 `dataframe`。
  - `kind='scatter'`：指定图的类型为散点图。
  - `color='r'`：设置图形的颜色为红色。
  - `size=10`：设置图形的大小。
  - `ratio=5`：设置主图与边缘图的比例为 5:1。
  - `space=1`：设置主图与边缘图之间的间距。

该图形同时展示了 `x` 和 `y` 的散点分布，以及两列数据的边缘分布（直方图）。

### 代码解释

以下代码使用 Seaborn 绘制了一个成对变量关系的图表（Pairplot）：

```python
sns.pairplot(data)
```

- `sns.pairplot(data)`：
  - `data`：输入的数据集，通常是一个 Pandas DataFrame。
  - 该函数会为数据集中每一对数值列绘制散点图，并在对角线上绘制每列的分布图（如直方图）。
  -对角线上：显示每个变量的单变量分布（默认是直方图）
  -非对角线上：显示每对变量之间的双变量关系（默认是散点图）

此图表非常适合用于探索数据集中变量之间的关系，尤其是在初步数据分析阶段。

类别散点图

### 代码解释

以下代码使用 Seaborn 绘制了一个类别散点图（Stripplot）：

```python
tips = sns.load_dataset('tips')
sns.stripplot(x='day', y='total_bill', data=tips)
```

- `sns.load_dataset('tips')`：加载 Seaborn 自带的 `tips` 数据集。
  - `tips` 数据集包含餐厅小费相关的数据，包括账单金额（`total_bill`）、小费金额（`tip`）、性别（`sex`）、是否吸烟（`smoker`）、用餐日期（`day`）、用餐时间（`time`）和用餐人数（`size`）。
- `sns.stripplot(...)`：绘制一个类别散点图。
  - `x='day'`：将用餐日期（`day`）作为横轴。
  - `y='total_bill'`：将账单金额（`total_bill`）作为纵轴。
  - `data=tips`：指定数据来源为 `tips` 数据集。

该图表展示了不同用餐日期下账单金额的分布情况，每个点代表一条记录。

x,y,hug标签,data,order,hue_order目标情况,jitte抖动程度

````python
tips=sns.load_dataset('tips')
sns.stripplot(x='day',y='total_bill',data=tips)
````

````python
sns.swarmplot(x='day',y='total_bill',data=tips)
````

### 代码解释

以下代码使用 Seaborn 绘制了一个分散型类别散点图（Swarmplot）：

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

- `sns.swarmplot(...)`：绘制一个分散型类别散点图。
  - `x='day'`：将用餐日期（`day`）作为横轴。
  - `y='total_bill'`：将账单金额（`total_bill`）作为纵轴。
  - `data=tips`：指定数据来源为 `tips` 数据集。

与 `stripplot` 不同，`swarmplot` 会调整点的位置以避免重叠，从而更清晰地展示数据分布。
该图表适合用于分析类别变量与数值变量之间的关系，同时保留数据点的分布信息。
