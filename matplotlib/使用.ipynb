```python
import matplotlib.pyplot as plt
import random

x = range(60)  # 确保 x 是 range 对象
y = [random.uniform(15, 18) for i in x]

z= [random.uniform(1, 3) for i in x]


plt.figure(figsize=(20, 8), dpi=100)
plt.plot(x, y,color='r',linestyle='--',label='上海')
plt.plot(x, z,label='北京')
# 正确生成 x 和 y 的刻度标签
x_ticks = ['11点({})分'.format(i) for i in x]
y_ticks = range(40)

# 设置 x 和 y 轴的刻度
plt.xticks(x[::5], x_ticks[::5])  # 每隔 5 个点设置一个刻度,x_ticks 中每隔 5 个点取一个标签，作为对应的刻度标签。
plt.yticks(y_ticks[::5])          # 每隔 5 个点设置一个刻度
plt.grid(True,linestyle="--",alpha=1)#这里的 True 表示启用网格线,通过 linestyle 参数指定网格线的样式,alpha 参数用于设置网格线的透明度，取值范围为 0 到 1。
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("温度变化图")
plt.savefig("./test.png")
plt.legend(loc='best')# 会为当前图表添加图例，并自动选择一个不会遮挡数据的最佳位置
plt.show()
```

plt.bar(x,width,align='center',**kwargs)  柱状图
x:传入的数据
 wigth:柱状图的宽度 
align:每个柱状图对齐方式{'center,'edge'},optional,default:'center'
 **kwargs:color:选择颜色

```python

# x:传入的数据
#  wigth:柱状图的宽度 
# align:每个柱状图对齐方式{'center,'edge'},optional,default:'center'
#  **kwargs:color:选择颜色
movie=['雷蛇','正义联盟','寻梦','追捕']
x=range(len(movie))
y=[75454,42564,65344,64213]
plt.figure(figsize=(20,8),dpi=100)
plt.bar(x,y,width=0.5,color=['b','r','g','y'])
plt.xticks(x,movie)
plt.grid(True,linestyle="--",alpha=0.5)#alpha透明度
plt.title("电影收入")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist([2,4,3,2,5,6], bins=None)#x数据 ，bin组距
```

`plt.hist([2,4,3,2,5,6], bins=None)` 绘制了一个直方图，其中 `[2,4,3,2,5,6]` 是数据列表。参数 `bins=None` 表示 Matplotlib 将自动选择适当的组距（bins）来划分数据范围。直方图用于显示数据的分布情况，每个柱子的高度表示该范围内数据点的数量。这种图表在分析数据分布和频率时非常有用。

```python

import matplotlib.pyplot as plt
plt.pie([50,50], labels=['数学', '语文'], autopct='%1.1f%%', colors=['red', 'blue'])
#x:数据
#label:每部分名称
#autopct:占比指定,用于在每个扇形上显示百分比值。'%1.1f%%' 是格式化字符串，表示保留一位小数并加上百分号。
#color:颜色
```
