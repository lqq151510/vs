`plt.subplot(111)` 创建一个子图对象。

- `111` 是一个三位数的代码，表示子图的网格布局。
- 第一个数字表示行数，第二个数字表示列数，第三个数字表示子图的索引。

在此例中，`111` 表示一个 1 行 1 列的网格布局，并选择第一个（也是唯一的）子图。

```python
x=np.linspace(0,2*np.pi,100)
x
```

`np.linspace(0, 2*np.pi, 100)` 用于生成一个等间距的数值数组。

- `0` 是起始值，表示数组从 0 开始。
- `2*np.pi` 是结束值，表示数组到 2π 结束。
- `100` 是生成的数值个数，表示数组中包含 100 个等间距的值。

此数组通常用于绘制周期函数（如正弦或余弦函数）的图形。

```python
y=np.sin(x)
y
```

```python
figure=plt.figure()#默认画板
axes1=plt.subplot(2,2,1)
plt.plot(x,y)#绘制线形图
axes2=plt.subplot(2,2,4)
plt.plot(x,y)#绘制线形图
axes1.plot(x,y)
axes2.plot(x,2*y)
```

`axes.axis([0, 2*np.pi+3, -1, 1])` 用于设置当前子图的坐标轴范围。

- `[0, 2*np.pi+3]` 指定 x 轴的范围，从 0 到 `2π+3`。
- `[-1, 1]` 指定 y 轴的范围，从 -1 到 1。

此方法可以通过传递一个列表 `[xmin, xmax, ymin, ymax]` 来一次性设置 x 和 y 轴的范围。

```python
plt.plot(x, y)
plt.xticks([0, np.pi/2, np.pi, np.pi*3/2, 2*np.pi], ['0', 'π/2', 'π', '3/2π', '2π'])
plt.yticks([-1, 0, 1], ['min', '0', 'max'])
```

```python
axes=plt.subplot(111)
axes.plot(x,y)
axes.set_xticks([0, np.pi/2, np.pi, np.pi*3/2, 2*np.pi])
axes.set_xticklabels(['0', 'π/2', 'π', '3/2π', '2π'])
axes.set_yticks([-1, 0, 1])
axes.set_yticklabels(['min', '0', 'max'])
```

```python
plt.plot(x,y)
plt.xlabel('x',color='red',fontsize=18)
plt.ylabel('y',rotation=0,color='red',fontsize=18)
```

`plt.xlabel('x', color='red', fontsize=18)` 用于设置 x 轴的标签。

- `'x'` 是标签的文本内容。
- `color='red'` 设置标签的颜色为红色。
- `fontsize=18` 设置标签的字体大小为 18。

`plt.ylabel('y', rotation=0)` 用于设置 y 轴的标签。

- `'y'` 是标签的文本内容。
- `rotation=0` 设置标签的旋转角度为 0 度，使其水平显示。

这些方法可以自定义坐标轴标签的外观，使图形更具可读性。

```python
axes=plt.subplot(111)
axes.plot(x,y)
axes.set_xlabel('x',color='red',fontsize=18)
axes.set_ylabel('y',rotation=0,color='red',fontsize=18)
```

```python
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1)
plt.title('sin(x)')
plt.show()
plt.plot(x,y2)
plt.title('cos(x)')
```

### 绘制正弦和余弦函数图像

1. `y1 = np.sin(x)`：计算数组 `x` 中每个值的正弦值，生成正弦函数的 y 值。
2. `y2 = np.cos(x)`：计算数组 `x` 中每个值的余弦值，生成余弦函数的 y 值。
3. `plt.plot(x, y1)`：绘制正弦函数的图像，x 轴为 `x`，y 轴为 `y1`。
4. `plt.title('sin(x)')`：为正弦函数图像添加标题。
5. `plt.show()`：显示正弦函数图像。
6. `plt.plot(x, y2)`：绘制余弦函数的图像，x 轴为 `x`，y 轴为 `y2`。
7. `plt.title('cos(x)')`：为余弦函数图像添加标题。

通过这段代码，可以分别绘制并显示正弦函数和余弦函数的图像。

```python
plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.legend() # 显示图例，用于标注每条曲线的含义
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(['sin(x)','cos(x)'])
```

```python
axes=plt.subplot(111)

axes.plot(x,y1,color='red')
axes.plot(x,y2)
axes.legend(['SIN(X)','COS(X)'])
```
