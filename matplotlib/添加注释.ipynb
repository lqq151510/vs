```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,2*np.pi,20)
y=np.sin(x)
```

```python
figure=plt.figure(figsize=(10,4),facecolor='lightgreen')  # Fixed color name
ax1=plt.subplot(1,2,1)
plt.plot(x,y)
plt.text(x=3,y=0,s='dance')# x,y表示文字在画布中的位置
ax2=plt.subplot(1,2,2)
plt.text(x=0.5,y=0.5,s='lucy',fontdict={'color':'green','fontsize':13})
figure.suptitle('My Figure Title',x=0.5,y=1)  #相对位置
```

### suptitle 和 title 的区别
1. **`suptitle`**:
   - 用于设置整个图形（`figure`）的标题。
   - 通过 `plt.figure().suptitle()` 或 `figure.suptitle()` 方法调用。
   - 标题的位置相对于整个图形，可以通过参数 `x` 和 `y` 调整。
   - 示例：`figure.suptitle('My Figure Title', x=0.5, y=1)`。

2. **`title`**:
   - 用于设置单个子图（`Axes`）的标题。
   - 通过 `plt.title()` 或 `ax.set_title()` 方法调用。
   - 标题的位置相对于子图，可以通过参数 `loc`（如 `'left'`, `'center'`, `'right'`）调整。
   - 示例：`ax1.set_title('Subplot Title', loc='center')`。

### 总结
- `suptitle` 是针对整个图形的标题，而 `title` 是针对单个子图的标题。
- 它们的作用范围不同，`suptitle` 更适合描述整个图形的主题，而 `title` 用于描述每个子图的内容。

```python
figure=plt.figure(figsize=(10,5))  # Fixed color name
ax1=plt.subplot(3,1,1)

ax2=plt.subplot(2,1,2)  # Create the second subplot
x=np.linspace(-np.pi,np.pi,50)
y=np.cos(x)
plt.annotate('max point', xy=(0, 1), xytext=(0, 0), arrowprops={'arrowstyle': '->'})
ax2.plot(x,y,color='red')
figure.text(x=0.5, y=0.55, s='liu',fontdict={'color':'green','fontsize':18                                                                                                                                     })  # Add text to the subplot
```

### `plt.annotate` 的功能和参数说明

`plt.annotate` 用于在图形中添加注释，标注特定的数据点。

#### 参数说明
- `s='max point'`：注释的文本内容，这里是 `'max point'`。
- `xy=(0, 1)`：被注释点的坐标，表示 x=0 和 y=1 的位置。
- `xytext=(0, 0)`：注释文本的位置，表示文本显示在 x=0 和 y=0 的位置。
- `arrowprops={'arrowstyle': '->'}`：设置箭头的样式，这里使用 `->` 表示简单的箭头。

#### 功能
这段代码在图形中标注了一个点 `(0, 1)`，并通过箭头将该点与注释文本 `'max point'` 连接起来。
注释文本显示在 `(0, 0)` 的位置，箭头样式为简单的直箭头。
