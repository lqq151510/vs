```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(-1,1,100)
f=lambda x:(1-x**2)**0.5
y=f(x)
```

```python
figure=plt.figure(figsize=(4,4),facecolor='green')#不会保存到本地
ax=plt.subplot(111)
plt.plot(x,y,color='red')
plt.plot(x,-y,color='red')
figure.savefig('my_ciecle.png',dpi=200,facecolor='red')
```
